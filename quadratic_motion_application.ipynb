### 放物運動の基礎

放物運動は、物体が地面に向かって投げられたり、あるいは上向きに投げられたりした際に、物体が描く軌跡のことです。放物運動は、二次関数を用いて表現されることが多く、その特徴的な軌道をグラフに描くことができます。

### 例題: 上向きに投げられた物体の運動

ある物体が高さ0mから秒速20mで上向きに投げられたとしましょう。この物体の運動は、次の二次関数で表されます。

$$
h(t) = -5t^2 + 20t
$$

ここで、$t$は時間（秒）、$h(t)$は物体の高さ（メートル）を表します。この式は、物体が上向きに投げられた後、どのように高さが変化していくかを示しています。

### 1. 物体が最高点に達する時間は何秒か？

最高点に達する時、物体の速度はゼロになります。このときの時間を求めるには、二次関数の速度（微分）を使いますが、簡単に言うと、放物運動では最高点の時間は次の式で求められます。

最高点に達する時間は、初速（$v_0$）を重力加速度（$g$）で割ったものとして求めることができます。具体的には次のように計算します。

$$
t = \frac{v_0}{|a|}
$$

ここで、$v_0 = 20$（初速）、$a = -10$（重力加速度の半分）なので、

$$
t = \frac{20}{10} = 2 \text{秒}
$$

したがって、物体が最高点に達するのは**2秒後**です。

### 2. 物体が地面に着地する時間は何秒か？

物体が地面に着地する時、$h(t) = 0$ になります。この時刻を求めるために、式 $h(t) = -5t^2 + 20t$ を 0 と置いて解きます。

$$
-5t^2 + 20t = 0
$$

これを因数分解すると、

$$
t(-5t + 20) = 0
$$

$t = 0$ または $t = 4$ となります。最初の $t = 0$ は物体が投げられた瞬間ですので、物体が地面に着地する時間は**4秒後**です。

### 3. 物体が最高点に達した時の高さは何メートルか？

最高点の高さは、先ほど求めた最高点の時間 $t = 2$ を式 $h(t) = -5t^2 + 20t$ に代入することで求められます。

$$
h(2) = -5(2)^2 + 20(2) = -5(4) + 40 = -20 + 40 = 20 \text{メートル}
$$

したがって、物体が最高点に達した時の高さは**20メートル**です。
