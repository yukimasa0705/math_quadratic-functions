## 二次関数の最大・最小を求める方法

二次関数 `y = ax^2 + bx + c` の最大・最小は、次の手順で求めます。

1. **頂点の座標を求める**  
   頂点の `x`-座標は、  
   `x_頂点 = -b / (2a)`  
   その `x`-座標を元に、対応する `y`-座標を求めます。  
   `y_頂点 = a * (-b / (2a))^2 + b * (-b / (2a)) + c`

2. **最大値・最小値の判定**  
   - `a > 0` の場合、放物線は上に開いており、頂点は最小値。
   - `a < 0` の場合、放物線は下に開いており、頂点は最大値。

### 例:  
関数 `y = 2x^2 - 4x + 1` の場合、  
1. 頂点の `x`-座標は `x_頂点 = 1`。  
2. `y_頂点 = -1` と求められるので、最小値は `y = -1` です。


## 二次関数の最大・最小を求める方法

二次関数 $y = ax^2 + bx + c$ の最大・最小は、次の手順で求めます。

### 1. 頂点の座標を求める
頂点の x-座標は次の式で求められます：
$$
x_{\text{頂点}} = \frac{-b}{2a}
$$

その x-座標を元に、対応する y-座標を求めます：
$$
y_{\text{頂点}} = a \left( \frac{-b}{2a} \right)^2 + b \left( \frac{-b}{2a} \right) + c
$$

### 2. 最大値・最小値の判定
- $a > 0$ の場合、放物線は上に開いており、頂点は最小値。
- $a < 0$ の場合、放物線は下に開いており、頂点は最大値。

### 例
関数 $y = 2x^2 - 4x + 1$ の場合、

頂点の x-座標は
$$
x_{\text{頂点}} = \frac{-(-4)}{2(2)} = 1
$$

y-座標は
$$
y_{\text{頂点}} = 2(1)^2 - 4(1) + 1 = -1
$$

よって、最小値は $y = -1$ です。

## 最適化問題：利益最大化

ある企業が製造する製品の利益を最大化したいと考えている。製品の利益は、製品の販売価格、製造コスト、および生産量に依存しており、次の二次関数で表されると仮定する。

利益関数：
$$
y = -2x^2 + 12x - 20
$$
ここで、$x$ は生産量（単位：千個）を表す。

### 最適化問題
この企業は、生産量 $x$ を決定することで、利益 $y$ を最大化したい。  
どの生産量が最も利益を最大化するかを求めなさい。

### 解法の手順
1. 利益関数 $y = -2x^2 + 12x - 20$ の頂点を求める。
2. 頂点の $x$ 座標を求めて、その値を生産量として最適化を図る。

#### 1. 頂点の $x$ 座標を求める
二次関数の頂点の $x$ 座標は次の式で求められます：
$$
x_{\text{頂点}} = \frac{-b}{2a}
$$
ここで、$a = -2$、$b = 12$ として、
$$
x_{\text{頂点}} = \frac{-12}{2 \times -2} = 3
$$

#### 2. 利益の最大値を求める
生産量 $x = 3$ を元に、利益関数に代入して最大利益を求めます：
$$
y_{\text{最大}} = -2(3)^2 + 12(3) - 20
$$
計算すると、
$$
y_{\text{最大}} = -18 + 36 - 20 = -2
$$

### 結論
最適な生産量は 3 千個で、最も高い利益（最大化された利益）は -2 千円となります。


```python
import numpy as np
import matplotlib.pyplot as plt

# 利益関数
def profit(x):
    return -2 * x**2 + 12 * x - 20

# xの範囲を設定（生産量は0から10千個くらいが適当）
x_vals = np.linspace(0, 6, 400)

# 利益の計算
y_vals = profit(x_vals)

# グラフの描画
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$y = -2x^2 + 12x - 20$', color='blue')
plt.axvline(x=3, color='red', linestyle='--', label=r'最適生産量 $x = 3$ 千個')
plt.axhline(y=-2, color='green', linestyle='--', label=r'最大利益 $y = -2$ 千円')

# グラフの詳細設定
plt.title('利益最大化のグラフ')
plt.xlabel('生産量 (千個)')
plt.ylabel('利益 (千円)')
plt.legend()
plt.grid(True)

# グラフ表示
plt.show()
```
